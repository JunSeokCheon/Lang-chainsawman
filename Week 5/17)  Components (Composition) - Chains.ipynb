# Chains
##### 체인은 LLM, 도구 또는 데이터 전처리 단계 등 일련의 호출을 나타낸다. 기본적으로는 LCEL을 사용한다.
##### LangChain에서는 지원하는 체인은 두 가지 유형이 있다.



*   LCEL로 구성된 체인 : LangChain은 더 높은 수준의 생성자 방법을 제공한다. 내부적으로는 LCEL을 사용하여 체인을 구축한다.
*   Legacy 체인 : Chain 클래스에서 서브클래스하여 구성된 체인이다. 해당 체인은 내부적으로 LCEL을 사용하지 않고 오히려 독립형 클래스이다.




## LCEL 체인


*   체인 생성자 : 체인의 생성자 함수이다. LCEL 실행 파일을 반환하는 모든 메서드이고 API 문서에 대한 링크도 제공한다.
*   함수 호출 : OpenAI 함수 호출이 필요한지 여부
*   기타 도구 : 다른 도구가 사용되는지 여부
*   사용 시기 : 언제 사용하는지에 대한 설명

## Legacy 체인

*   체인 : 체인의 이름 또는 생성자 메서드의 이름이다. 생성자 메서드인 경우 Chain 하위 클래스가 반환된다.
*   함수 호출 : OpenAI 함수 호출이 필요한지 여부
*   기타 도구 : 다른 도구가 사용되는지 여부
*   사용 시기 : 언제 사용하는지에 대한 설명

##### Legacy 체인은 너무 많은 관계로 링크로 대체한다.

##### LLMChain 예제

```python
from langchain.chains import LLMChain
from langchain_community.llms import OpenAI
from langchain_core.prompts import PromptTemplate
prompt_template = "Tell me a {adjective} joke"
prompt = PromptTemplate(
    input_variables=["adjective"], template=prompt_template
)
llm = LLMChain(llm=OpenAI(), prompt=prompt)
```
